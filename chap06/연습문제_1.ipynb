## 연습문제 1: Person 클래스

이 파일은 Person 클래스에 대한 연습문제를 다루는 노트입니다.

학번: 202110763

* 연습문제 6-8. 위의 Person 클래스를 수정하여 Employee 클래스를 포함하여 새로운 Person 객체가 생성될 때마다 ID가 1001부터 순차적으로 자동 할당되도록 하라. 새로운 Employee 객체가 생성되고 getId()를 호출할 때 ID가 중복되지 않아야 한다.


```python
class Person:
    _id_counter = 1000  # 클래스 변수로 ID 카운터 관리
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        Person._id_counter += 1  # 새로운 객체 생성 시 ID 증가
        self._id = Person._id_counter
    
    def getId(self):
        print(f"ID: {self._id}")
        return self._id
    
    def getAge(self):
        print(f"나이: {self.age}")
        return self.age
    
    def getName(self):
        print(f"이름: {self.name}")
        return self.name

class Employee(Person):
    def __init__(self, name, age, department):
        super().__init__(name, age)  # 부모 클래스 생성자 호출
        self.department = department
    
    def getDepartment(self):
        print(f"부서: {self.department}")
        return self.department
    
    def getInfo(self):
        print(f"직원 정보 - ID: {self._id}, 이름: {self.name}, 나이: {self.age}, 부서: {self.department}")

# 테스트
print("=== Person 객체 생성 ===")
person1 = Person("김철수", 25)
person2 = Person("이영희", 30)

print("Person1 정보:")
person1.getId()
person1.getName()
person1.getAge()

print("\nPerson2 정보:")
person2.getId()
person2.getName()
person2.getAge()

print("\n=== Employee 객체 생성 ===")
employee1 = Employee("박민수", 28, "개발팀")
employee2 = Employee("정수진", 32, "마케팅팀")

print("Employee1 정보:")
employee1.getInfo()

print("\nEmployee2 정보:")
employee2.getInfo()

print(f"\n=== ID 중복 확인 ===")
print(f"Person1 ID: {person1._id}")
print(f"Person2 ID: {person2._id}")
print(f"Employee1 ID: {employee1._id}")
print(f"Employee2 ID: {employee2._id}")
print("모든 ID가 순차적으로 할당되었습니다!")
```
